```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

$$ \alpha = \frac{mgh}{kT} $$

```python
alpha1 = 2  # Dominan los efectos mecánicos 
alpha2 = 0.5    # Dominan los efectos térmicos

N = 2000
points1 = np.zeros((N,3))
points2 = points1.copy() 

for i in range(N):
    points1[i] = [ np.random.rand(), np.random.rand(), np.random.exponential(scale=1/alpha1) ]
    points2[i] = [ np.random.rand(), np.random.rand(), np.random.exponential(scale=1/alpha2) ]
```

```python
fig = plt.figure(figsize=(10,8))
ax  = fig.add_subplot(1,2,1, projection = '3d',elev = 8, azim = 45)
ax1 = fig.add_subplot(1,2,2, projection = '3d',elev = 8, azim = 45)

ax.set_title(r'$kT \ll mgh$',fontsize=15)
ax.set_zlim(0,3)
ax1.set_title(r'$kT \gg mgh$',fontsize=15)
ax1.set_zlim(0,3)
ax.scatter(points1[:,0],points1[:,1],points1[:,2])
ax1.scatter(points2[:,0],points2[:,1],points2[:,2]);
```
